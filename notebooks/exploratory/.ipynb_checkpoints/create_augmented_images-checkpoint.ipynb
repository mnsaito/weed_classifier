```python
# Standard Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
import os

# Modeling Imports
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_path = os.path.join(os.pardir, os.pardir, 'data', 'train', 'perennials')
data_file = 'delphinium_exaltatum_mo_0.jpg'
data_img = mpimg.imread(os.path.join(data_path, data_file))

store_path = os.path.join(os.pardir, os.pardir, 'reports', 'images')
original = os.path.join(store_path, 'original_plant')

fig = plt.figure(figsize = (20, 8))
plt.imshow(data_img)
fig.suptitle('Original', fontsize = 18)
plt.savefig(original)
```

```python
datagen = ImageDataGenerator()
translate = datagen.apply_transform(x=data_img, transform_parameters={'theta':30})
rotate_30 = os.path.join(store_path, 'rotate_30')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Rotate +30', fontsize = 18)
plt.savefig(rotate_30)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'theta':-30})
rotate_neg_30 = os.path.join(store_path, 'rotate_neg_30')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Rotate -30', fontsize = 18)
plt.savefig(rotate_neg_30)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'flip_horizontal':180})
flip = os.path.join(store_path, 'flip')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Horizontal Flip', fontsize = 18)
plt.savefig(flip)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'zx':0.7, 'zy':0.7})
zoom_in = os.path.join(store_path, 'zoom_in')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Zoom In', fontsize = 18)
plt.savefig(zoom_in)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'zx':1.3, 'zy':1.3})
zoom_out = os.path.join(store_path, 'zoom_out')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Zoom Out', fontsize = 18)
plt.savefig(zoom_out)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'ty':100})
shift_left = os.path.join(store_path, 'shift_left')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Shift Left', fontsize = 18)
plt.savefig(shift_left)
```

```python
translate = datagen.apply_transform(x=data_img, transform_parameters={'tx':-100})
shift_down = os.path.join(store_path, 'shift_down')

fig = plt.figure(figsize = (20, 8))
plt.imshow(translate)
fig.suptitle('Shift Down', fontsize = 18)
plt.savefig(shift_down)
```
