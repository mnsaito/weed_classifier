```python
# Standard Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
import os

# Modeling Imports
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Obtain original photo
data_path = os.path.join(os.pardir, os.pardir, 'data', 'train', 'perennials')
data_file = 'delphinium_exaltatum_mo_0.jpg'
data_img = mpimg.imread(os.path.join(data_path, data_file))

datagen = ImageDataGenerator()

# Rotate 30 degrees
rotate_30 = datagen.apply_transform(x=data_img, transform_parameters={'theta':30})

# Rotate -30 degrees
rotate_neg_30 = datagen.apply_transform(x=data_img, transform_parameters={'theta':-30})

store_path = os.path.join(os.pardir, os.pardir, 'reports', 'images')
augment_1 = os.path.join(store_path, 'augment_1')
augment_2 = os.path.join(store_path, 'augment_2')
augment_3 = os.path.join(store_path, 'augment_3')

# Store photos
plt.figure(figsize = (20, 8))
ax = plt.subplot(1, 3, 1)
ax.set_title('Original', fontdict={'size': 18})
plt.imshow(data_img)
ax = plt.subplot(1, 3, 2)
ax.set_title('Rotate 30 Degrees', fontdict = {'size': 18})
plt.imshow(rotate_30);
ax = plt.subplot(1, 3, 3)
ax.set_title('Rotate -30 Degrees', fontdict = {'size': 18})
plt.imshow(rotate_neg_30);
plt.savefig(augment_1)
```

```python
# Horizontal Flip
flip = datagen.apply_transform(x=data_img, transform_parameters={'flip_horizontal':180})

# Zoom In
zoom_in = datagen.apply_transform(x=data_img, transform_parameters={'zx':0.7, 'zy':0.7})

# Zoom Out
zoom_out = datagen.apply_transform(x=data_img, transform_parameters={'zx':1.3, 'zy':1.3})

# Store photos
plt.figure(figsize = (20, 8))
ax = plt.subplot(1, 3, 1)
ax.set_title('Horizontal Flip', fontdict={'size': 18})
plt.imshow(flip)
ax = plt.subplot(1, 3, 2)
ax.set_title('Zoom In', fontdict = {'size': 18})
plt.imshow(zoom_in);
ax = plt.subplot(1, 3, 3)
ax.set_title('Zoom Out', fontdict = {'size': 18})
plt.imshow(zoom_out);
plt.savefig(augment_2)
```

```python
# Shift Left
shift_left = datagen.apply_transform(x=data_img, transform_parameters={'ty':100})

# Shift Down
shift_down = datagen.apply_transform(x=data_img, transform_parameters={'tx':-100})

# Adjust Brightness
brightness = datagen.apply_transform(x=data_img, transform_parameters={'brightness':0.006})

# Store photos
plt.figure(figsize = (20, 8))
ax = plt.subplot(1, 3, 1)
ax.set_title('Shift Left', fontdict={'size': 18})
plt.imshow(shift_left)
ax = plt.subplot(1, 3, 2)
ax.set_title('Shift Down', fontdict = {'size': 18})
plt.imshow(shift_down);
ax = plt.subplot(1, 3, 3)
ax.set_title('Adjust Brightness', fontdict = {'size': 18})
plt.imshow(brightness);
plt.savefig(augment_3)
```
